```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
start = '2014-01-01'
end = '2024-01-01'

stock = 'PEP'

df = yf.download(stock, start=start, end=end)

# Save the data to a CSV file
#data.to_csv(f'{stock}_5_data.csv')
```

```python
import pandas as pd



# Forward fill missing dates
df = df.resample('D').ffill()

# Sort the DataFrame in the latest to earliest order
df = df.sort_index(ascending=False)

print(df)
```
